Occupations
Step 1. Import the necessary libraries
Step 2. Import the dataset from this address.
Step 3. Assign it to a variable called users
Step 4. Discover what is the mean age per occupation
Step 5. Discover the Male ratio per occupation and sort it from the most to the least
Step 6. For each occupation, calculate the minimum and maximum ages
Step 7. For each combination of occupation and sex, calculate the mean age
Step 8. For each occupation present the percentage of women and men

```python
import pandas as pd
import numpy as np

# Step 2:
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user'

# Step 3: 
users = pd.read_csv(url, sep='|')

users.head()
```

```python
# Step 4:
mean_age_per_occupation = users.groupby('occupation')['age'].mean().reset_index()

print("Mean Age per Occupation:")
display(mean_age_per_occupation)
```

```python
# Step 5: 
male_ratio_per_occupation = (users[users['gender'] == 'M'].groupby('occupation').size() / 
                             users.groupby('occupation').size()).sort_values(ascending=False).reset_index(name='Male Ratio')

print("Male Ratio per Occupation (Sorted):")
display(male_ratio_per_occupation)
```

```python
# Step 6:
min_max_age_per_occupation = users.groupby('occupation')['age'].agg(['min', 'max']).reset_index()

print("Minimum and Maximum Age per Occupation:")
display(min_max_age_per_occupation)
```

```python
# Step 7: 
mean_age_per_occupation_sex = users.groupby(['occupation', 'gender'])['age'].mean().unstack().reset_index()

print("Mean Age per Occupation and Gender:")
display(mean_age_per_occupation_sex)
```

```python
# Step 8:
gender_percentage_per_occupation = (users.groupby(['occupation', 'gender']).size().unstack().apply(lambda x: (x / x.sum()) * 100, axis=1).reset_index())

print("Gender Percentage per Occupation:")
display(gender_percentage_per_occupation)
```

```python
# Step 3

data = pd.read_csv('wind.txt', delim_whitespace=True)
display(data.head())
# Create a DateTime index from 'Yr', 'Mo', and 'Dy'
data['Date'] = pd.to_datetime(data[['Yr', 'Mo', 'Dy']].astype(str).agg('-'.join, axis=1), format='%y-%m-%d')
data = data.set_index('Date')
data = data.drop(columns=['Yr', 'Mo', 'Dy'])
display(data.head())
```

```python
# Step 4
def fix_year(x):
    if x.year > 2060:
        return x.replace(year=x.year - 100)
    return x

data.index = data.index.map(fix_year)
display(data.head())  
```

```python
# Step 11
january_data = data[data.index.month == 1]
january_avg = january_data.groupby(january_data.index.year).mean()
display(january_avg)
```

# Question 6

Create a line plot showing the number of marriages and divorces per capita in the 
U.S. between 1867 and 2014. Label both lines and show the legend.
Don't forget to label your axes!

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('us-marriages-divorces-1867-2014.csv')

plt.figure(figsize=(10, 6))
plt.plot(data['Year'], data['Marriages_per_1000'], label='Marriages per Capita')
plt.plot(data['Year'], data['Divorces_per_1000'], label='Divorces per Capita')

plt.xlabel('Year')
plt.ylabel('Per Capita Rate')
plt.title('Marriages and Divorces per Capita in the U.S. (1867-2014)')
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv('roman-emperor-reigns.csv')

data = data[data['Cause_of_Death'] == 'Assassinated']


fig, ax = plt.subplots(figsize=(20, 16))
patches, texts = ax.pie(data['Length_of_Reign'], labels=None, startangle=140)


ax.legend(patches, data['Emperor'], loc="center left", bbox_to_anchor=(1, 0.5), fontsize='small')


for i, p in enumerate(patches):
    ang = (p.theta2 - p.theta1) / 2. + p.theta1
    y = np.sin(np.deg2rad(ang))
    x = np.cos(np.deg2rad(ang))
    horizontalalignment = {-1: 'right', 1: 'left'}[int(np.sign(x))]
    connectionstyle = "angle,angleA=0,angleB={}".format(ang)

    
    if i == 10:  
        ax.annotate(f'{data["Length_of_Reign"].iloc[i] / data["Length_of_Reign"].sum() * 100:.1f}%',
                    xy=(x, y), xytext=(3.5*np.sign(x), 3.5*y),  # Increase distance for this label
                    arrowprops=dict(arrowstyle="->", color='black', connectionstyle="arc3,rad=0.4"),
                    bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="black", lw=1),
                    fontsize=10, color='black', ha=horizontalalignment, va='center')
    elif i == 12:  
        ax.annotate(f'{data["Length_of_Reign"].iloc[i] / data["Length_of_Reign"].sum() * 100:.1f}%',
                    xy=(x, y), xytext=(3.0*np.sign(x), 2.5*y),  # Slightly different positioning
                    arrowprops=dict(arrowstyle="->", color='black', connectionstyle="arc3,rad=0.3"),
                    bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="black", lw=1),
                    fontsize=10, color='black', ha=horizontalalignment, va='center')
    else:
        ax.annotate(f'{data["Length_of_Reign"].iloc[i] / data["Length_of_Reign"].sum() * 100:.1f}%',
                    xy=(x, y), xytext=(2.7*np.sign(x), 2.7*y),  # General increase for all others
                    arrowprops=dict(arrowstyle="->", color='black', connectionstyle=connectionstyle),
                    bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="black", lw=1),
                    fontsize=10, color='black', ha=horizontalalignment, va='center')

plt.show()
```
